```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/UdeshaHerath/FYP/refs/heads/main/V13July_avg.csv')
df.head()
```

```python
# Check for missing values
print(df.isnull().sum())
```

No missing data


```python
Filterd_df = df[['Date_Time','Active_Power','Wind_Speed','Nacelle Temperature','Ambient_Temperature','Gear_Oil_Tempeature','Gearbox Bearing Temperature','Generator_RPM']]
Filterd_df=Filterd_df[0:2456]
Filterd_df.tail(20)
```

```python
Filterd_df.info()
Filterd_df.describe()
```

```python
# prompt: scatter plot Wind_Speed vs Active_Power wide plot

import matplotlib.pyplot as plt

plt.figure(figsize=(18, 6))  # Set the figure size for a wider plot
plt.scatter(Filterd_df['Wind_Speed'], Filterd_df['Active_Power'], alpha=0.5)
plt.xlabel('Wind Speed')
plt.ylabel('Active Power')
plt.title('Scatter plot of Wind Speed vs Active Power')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(18, 6))  # Set the figure size for a wider plot
plt.scatter(Filterd_df['Generator_RPM'], Filterd_df['Active_Power'], alpha=0.5)
plt.xlabel('Wind Speed')
plt.ylabel('Active Power')
plt.title('Scatter plot of Generator_RPM vs Active Power')
plt.grid(True)
plt.show()
```
